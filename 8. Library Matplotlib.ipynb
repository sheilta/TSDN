# Visualisasi Data  


Untuk melakukan visualisasi data, digunakan package matplotlib. Dalam notebook ini akan dipelajari: diagram garis, diagram batang, histogram, boxplot, dan pie chart

## Grafik Garis dan Scatter 

```python
# Import subpackage pyplot sebagai plt
import matplotlib.pyplot as plt
```

```python
# Buat variabel tahun, dari 2005 hingga 2017
tahun = list(range(2007, 2018))
tahun
```

```python
# Data populasi dunia dalam milyar, sumber: https://data.worldbank.org/indicator/SP.POP.TOTL?end=2017&start=1960&view=chart
pop = [70.119, 71.034, 71.957, 72.885, 73.811, 74.755, 75.718, 76.689, 77.661, 78.640, 79.637]
```

```python
# buat diagram garis dengan plt.plot(x,y) 
plt.plot(tahun, pop)

# Tampilkan plot dengan plt.show()
plt.show()
```

```python
# mengubah warna dan tipe garis, dan memberikan tanda
plt.plot(tahun, pop,"ro--")


# Tampilkan plot dengan plt.show()
plt.show()
```

```python
# Buat diagram scatter
plt.scatter(tahun, pop)

# Annotate
plt.annotate(xy=[2014, 76], s = "Lihat tahun ini")

plt.show()
```

**Bagaimana korelasinya?**

```python
# Populasi wanita sedunia dalam milyar, sumber: https://data.worldbank.org/indicator/SP.POP.TOTL.FE.IN

pop_wanita = [34.762, 35.213, 35.668, 36.126, 36.582, 37.048, 37.522, 38.001, 38.482, 38.965, 39.449]
pop_laki2 = [35.330, 35.793, 36.261, 36.731, 37.201, 37.680, 38.168, 38.659, 39.151, 39.646, 40.140]
```

```python
# Gambarkan plot tahun terhadap populasi laki2 dan populasi wanita

plt.plot(tahun, pop_laki2, "r>-", label = "Populasi Laki2", linewidth = 1)
plt.plot(tahun, pop_wanita, "g>-", label = "Populasi Wanita", linewidth = 1)

plt.title("Jumlah Populasi Berdasar Gender")
plt.xlabel("Tahun")
plt.ylabel("Jumlah")

plt.legend(loc="upper left")

plt.grid(True, color="k")

plt.show()
```

## Grafik Batang

```python
# Gambarkan grafik batang

perusahaan = ["GOOG", "AAPL", "FB", "MSFT", "BABA", "AMZN"]
eps = [23.16, 11.04, 6.46, 2.13, 3.37, 12.63]
```

```python
# Gunakan plt.bar(x,y,color=...) untuk membuat grafik batang
plt.bar(perusahaan, eps, color="c")

plt.title("EPS Perusahaan")
plt.xlabel("Kode Finansial Perusahaan")
plt.ylabel("Earning Per Share (EPS)")

plt.show()
```

```python
# Variabel saham PE RATIO 
pe_ratio = [49.62, 20.27, 24.33, 52.04, 44.89, 147.65]
```

```python
# Dalam satu grafik, gambarkan eps dan pe_ratio dengan perusahaan sebagai axis
plt.bar(perusahaan, eps, label ="EPS", color="c")
plt.bar(perusahaan, pe_ratio, label = "PE Ratio", color="r")

plt.title("EPS Perusahaan")
plt.xlabel("Kode Finansial Perusahaan")

plt.legend()

plt.show()
```

**Apa yang menjadi masalah dalam visualisasi tersebut?**

```python
import numpy as np
```

```python
xpos = np.arange(len(perusahaan))
xpos
```

```python
# Gunakan plt.bar(x,y,color=...) untuk membuat grafik batang
plt.bar(xpos-0.2, eps, label ="EPS", color="c", width = 0.4)
plt.bar(xpos+0.2, pe_ratio, label = "PE Ratio", color="r", width = 0.4)

plt.xticks(xpos, perusahaan)


plt.title("EPS Perusahaan")
plt.xlabel("Kode Finansial Perusahaan")

plt.legend()

plt.show()
```

## Histogram

```python
import pandas as pd
```

```python
superstore = pd.read_excel("Superstore.xls")
superstore.head()
```

```python
plt.hist(superstore["Discount"], bins = 5)

plt.title("Histogram Diskon")
plt.xlabel("Nilai Diskon")
plt.ylabel("Frekuensi")

# Menyimpan dalam bentuk png
plt.savefig("Histogram_diskon.png")
```

```python
# Fungsi histogram dapat dipanggil dalam obyek 
superstore["Discount"].hist(bins=5)
plt.show()
```

## Grafik Frekuensi

```python
# Hitung frekuensi kemunculan provinsi dalam superstore menggunakan .value_count()
superstore["State"].value_counts()
```

```python
# Gambarkan grafik batang frekuensi
superstore["State"].value_counts().plot("bar", color="g")

plt.title("Grafik Persebaran Provinsi")
plt.xlabel("Provinsi")
plt.ylabel("Frekuensi")

plt.show()
```

## Boxplot

```python
plt.boxplot(superstore["Profit"], vert = False)
plt.show()
```

```python
fig, axs = plt.subplots(2,1)
axs[0].boxplot(superstore["Profit"], vert = False)
axs[1].boxplot(superstore["Discount"], vert = False)
plt.show()
 
```

```python
kategori = pd.DataFrame(superstore["Sub-Category"].value_counts()) 
kategori.head()
```

## Pie Chart

```python
indeks = np.array(kategori.index) 
indeks
```

```python
explode = ()

plt.pie(kategori["Sub-Category"], labels = indeks, autopct='%1.0f%%', startangle = 150)
plt.axis('equal')

plt.show()
```
